### Praktikum 01: Inheritance Dasar

Berikut adalah contoh sederhana dalam python yang mendemonstrasikan proses pewarisan (inheritance). Kelas diagram memperikan contoh proses pewarisan.

image.png

Berikut adalah penjelasan untuk kelas diagram dari Praktikum 1,
1. Kelas Kenadaraan

o	Atribut Publik (+):

-	`merk`: str: Menyimpan merk kendaraan.

o	Metode Publik (+):

- `__init__` (merk: str) : void: Konstruktor untuk inisialisasi merk.

-	`mulai_mesin()` : void: Metode untuk simulasi menyalakan mesin.

2. Kelas Mobil (Mewarisi kendaraan)

o	Atribut Publik (+):

-	`warna`: str: Menyimpan warna mobil (atribut spesifik Mobil).

-	`merk`: str: Diwarisi dari Kendaraan, namun dalam contoh kode Praktikum 1 ini, nilainya diatur ulang secara eksplisit di dalam konstruktor Mobil.

o	Metode Publik (+):

- `__init__` (merk: str, warna: str) : void: Konstruktor untuk Mobil, menginisialisasi merk dan warna. (Belum menggunakan super() di praktikum ini).

- `info_mobil()` : void: Metode spesifik untuk menampilkan info mobil. (Metode mulai_mesin diwarisi dari Kendaraan).

Diagram ini menunjukkan hubungan pewarisan dasar dimana kelas Mobil mewarisi (<|--) dari kelas Kendaraan. Objek Mobil dapat mengakses metode mulai_mesin yang didefinisikan di Kendaraan.
Kemudian untuk kode praktikum dalam python dapat dilihat sebagai berikut:


```
# Kelas Index
class Kendaraan:
  def _init_(self, merk):
    self.merk = merk
  def mulai_mesin(self):
    print(f"Mesin kendaraan {self.merk} dinyalakan")

# Kelas Anak (mewarisi dari Kendaraan)
class Mobil(Kendaraan):
  def __init__(self,merk, warna):
    self.merk = merk
    self.warna = warna

  def info_mobil(self):
    print(f"Ini adalah mobil {self.merk} bewarna {self.warna}.")

# Kode utama
if __name__ == "__main__":
  mobil_tesla = Mobil("Tesla Model S", "merah")

  mobil_tesla.info_mobil()

  mobil_tesla.mulai_mesin()

  print(f"Merk Mobil: {mobil_tesla.merk}")
```

### Praktikum 02 : Menggunakan (super) dalam konstruktor


Tujuan: Menggunakan `super()._init_()` untuk inisialisasi atribut kelas induk `(Person)` dari dalam konstruktor kelas anak `(Student)`.
image.png

1. Kelas Person

o	Atribut Publik (+): `nama, usia.`

o	Metode Publik (+): `__init__(nama, usia), perkenalkan_diri().`

2. Kelas Student(Mewakili dari Person)

o	Atribut Publik (+): `student_id, jurusan.` (Atribut nama dan usia diwarisi).

o	Metode  Publik  (+): `__init__(nama,  usia,  student_id,  jurusan)`  (memanggil
`super().__init__()`), `info_akademik(), perkenalkan_diri()` (meng-override metode induk).

Diagram ini menunjukkan kelas Student yang mewarisi dari Person. Fokus utama praktikum ini adalah pada metode  init  di Student yang menggunakan super(). init () untuk memanggil konstruktor Person, memastikan inisialisasi yang benar dari atribut nama dan usia sebelum menambahkan atribut spesifik Student.

Kode Python:

```
# Kelas Induk
class Person:
  def __init__(self, nama, usia):
    print(f"(Memanggil __init__ Person untuk '{nama}')")
    self.nama = nama
    self.usia = usia

  def perkenalkan_diri(self):
    print(f"Halo, Nama saya {self.nama}, usia saya {self.usia} tahun.")

# Kelas Anak (mewarisi dari Person)
class Student(Person):
  def __init__(self, nama, usia, student_id, jurusan):
    print(f"(Memanggil __init__ student untuk '{nama})")
    #memanggil __init__ dari kelas Person untuk inisialisasi nama dan usia

    super().__init__(nama, usia)
    #menambahkan atribut khsus Student
    self.student_id = student_id
    self.jurusan = jurusan
    print(f"(Inisialisasi attribut Student selesai untuk ('{nama}'))")

  def info_akademik(self):
    print(f"ID Mahasiswa: {self.student_id}")
    print(f"Jurusan: {self.jurusan}")

  # Override perkenalkan_diri untuk menampilkan info student
  def perkenalkan_diri(self):
    super().perkenalkan_diri() # Panggil versi person
    print(f"Saya adalah Mahasiswa dengan ID {self.student_id}, jurusan {self.jurusan}.")

# Kode Utama
if __name__ == "__main__":
  # Membuat Objek person
  dosen = Person("Pak Anton", 45)
  print("-"*20)
  dosen.perkenalkan_diri()
  print("\n"+ "=" * 30 + "\n")

  # Membuat Objek student
  mahasiswa = Student("Azka", 19, "4.33.24.1.04", "Teknik Elektro")
  print("-"*20)
  mahasiswa.perkenalkan_diri() #panggil override di Student
  print("-"*20)
  mahasiswa.info_akademik

  #cek attribut yang diwarisi
  print(f"\nUsia mahasiswa {mahasiswa.nama}: {mahasiswa.usia}") # Usia diwarisi oleh person
```

### Praktikum 03: Method Overriding dan super()

Tujuan: Mengganti (override) metode dari kelas induk di kelas anak dan menggunakan super() untuk memanggil implementasi asli dari induk.

image.png

Penjelasan Kelas Diagram:
1.Kelas Hewan:

o	Atribut Publik (+):

- `nama`: str: Menyimpan nama hewan.

o	Metode Publik (+):

- `__init__`(nama: str) : void: Konstruktor untuk inisialisasi nama.

- `bersuara()` : void: Metode dasar untuk hewan bersuara.

- `tampilkan_info()` : void: Metode dasar untuk menampilkan info hewan.

2.Kelas Kucing (Mewarisi dari Hewan):

o	Atribut Publik (+):

-	ras: str: Menyimpan ras kucing. (Atribut nama diwarisi).

o	Metode Publik (+):
- `__init__` (nama:	str,	ras:	str)	:	void:	Konstruktor	Kucing.	Memanggil
`super().__init__(nama)` untuk inisialisasi nama dari Hewan, lalu inisialisasi ras.

- `bersuara()` : void: Override metode bersuara dari Hewan untuk memberikan implementasi suara kucing yang spesifik ("Meow!").

- `tampilkan_info()` : void: Override metode tampilkan_info dari Hewan. Metode ini memanggil `super()`.`tampilkan_info()` untuk menjalankan implementasi asli dari Hewan, kemudian menambahkan informasi spesifik tentang ras kucing.



```
#Kelas Induk

class Hewan:
  def __init__(self, nama):
    self.nama = nama

  def bersuara(self):
    print(f"{self.nama} mengeluarkan suara generik")

  def tampilkan_info(self):
    print(f"ini adalah hewan bernama {self.nama}")

# Kelas Anak
class Kucing(Hewan):
  def __init__(self, nama, ras):
    super().__init__(nama) # Panggil init induk
    self.ras = ras

  # Method Overridng untuk bersuara
  def bersuara(self):
    print(f"{self.nama} (Kucing) mengeong: Meow!")

  # Methode Overriding untuk tampilkan_onfo , memanggil versi induk
  def tampilkan_info(self):
    super().tampilkan_info() #pnggil implementasi dari Hewan
    print(f"Ini adalah kucong ras {self.ras}") #tambahan info spesifik

# Kode utama
if __name__ == "__main__":
  hewan_umum = Hewan("Makhluk")
  kucing_persia = Kucing("Puspus", "Persia")

  print("Info hewan Umum:")
  hewan_umum.tampilkan_info()
  hewan_umum.bersuara()
  print("-"*20)

  print("Info kucing persia:")
  kucing_persia.tampilkan_info() # Akan memanggil versi Kucing
  kucing_persia.bersuara() # Akan memanggil versi Kucing
```
